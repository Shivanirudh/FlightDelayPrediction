## **Key Data**

```python
airports = ['ATL', 'CLT', 'DEN', 'DFW', 'EWR', 'IAH', 'JFK', 'LAS', 'LAX', 'MCO', 'MIA', 'ORD', 'PHX', 'SEA', 'SFO']
```

```python
f = pd.read_csv('/content/drive/My Drive/SF/2016/On_Time_On_Time_Performance_2016_1/On_Time_On_Time_Performance_2016_1.csv')
keys = f.Dest.unique()
```

```python
ID_dict = {}
for base in airports:
    if base in keys:
        ixs = f[f['Dest'] == base].index.tolist()
        ID_dict[base] = f.iloc[ixs[0]]['DestAirportID']
```

```python
ID_list = []
for i in ID_dict.keys():
    ID_list.append(ID_dict[i])
```

## **Weather Data** 

```python
meta = ['mintempC', 'maxtempF', 'sunHour', 'mintempF', 'maxtempC','totalSnow_cm', 'date', 'uvIndex']
```

```python
base_path = "/content/drive/My Drive/SF/weather"
```

```python
cols = ['windspeedKmph', 'winddirDegree', 'weatherCode', 'precipMM', 'visibility', 'pressure', 'cloudcover', 'DewPointF', 'WindGustKmph', 'tempF', 'WindChillF', 'humidity', 'date', 'time']
```

```python
weather_data = pd.DataFrame()
```

```python
for dir in airports:
    dir_df = pd.DataFrame()
    for i in range(1, 13):
        dir_path = os.path.join(base_path, dir)
        file_path = '2016-' + str(i) + '.json' 
        path = os.path.join(dir_path, file_path)
        file = open(path, "r+")
        new_data = json.load(file)
        df = pd.DataFrame(new_data)
        hourly_data = pd.json_normalize(data = df['data']['weather'], record_path = 'hourly', meta = meta)
        astronomy_data = pd.json_normalize(data = df['data']['weather'], record_path = 'astronomy', meta = meta)
        new_df = pd.merge(hourly_data, astronomy_data, on = meta)
        new_df = new_df[cols]
        dir_df = pd.concat([dir_df, new_df])
    for i in range(1, 13):
        dir_path = os.path.join(base_path, dir)
        file_path = '2017-' + str(i) + '.json' 
        path = os.path.join(dir_path, file_path)
        file = open(path, "r+")
        new_data = json.load(file)
        df = pd.DataFrame(new_data)
        hourly_data = pd.json_normalize(data = df['data']['weather'], record_path = 'hourly', meta = meta)
        astronomy_data = pd.json_normalize(data = df['data']['weather'], record_path = 'astronomy', meta = meta)
        new_df = pd.merge(hourly_data, astronomy_data, on = meta)
        new_df = new_df[cols]
        dir_df = pd.concat([dir_df, new_df])
    dir_df['airport'] = ID_dict[dir]
    weather_data = pd.concat([weather_data, dir_df])
```

## **Combining Data**

```python
flights_data = pd.read_csv('/content/flight_data.csv')
weather_data = pd.read_csv('/content/weather_data.csv')
```

```python
flights_data.isnull()
weather_data.isnull()
```

```python
f_df = pd.DataFrame()
for i in ID_list:
    f_df = pd.concat([f_df, flights_data[flights_data['OriginAirportID'] == i]])
```

```python
new_f_df = pd.DataFrame()
for i in ID_list:
    new_f_df = pd.concat([new_f_df, f_df[f_df['DestAirportID'] == i]])
```

```python
new_f_df
```

```python
flights_df = pd.merge(new_f_df, weather_data, left_on = ['OriginAirportID', 'FlightDate', 'RoundDepTime'], right_on = ['airport', 'date', 'time'], how = 'inner')
```

```python
flights_df.drop(['RoundDepTime'], axis = 1, inplace = True)
```
